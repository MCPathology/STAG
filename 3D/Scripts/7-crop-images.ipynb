```python
import pandas as pd
from PIL import Image
import os
from pathlib import Path

from PIL import Image
Image.MAX_IMAGE_PIXELS = None # <-- 添加这一行

import pandas as pd
# ... 你的其他代码 ...

# --- 1. 路径设置 ---
# !!! 重要: 如果你的 notebook 文件和 data 文件夹在同一级，请将下面这行改为 base_data_folder = Path('data')
base_data_folder = Path('../data') 

datasets = {
    'skin': base_data_folder / 'skin',
}

CROP_SIZE = 224

# --- 2. 主处理函数 ---
def process_dataset(name, base_path):
    print(f"--- 开始处理数据集: {name} ---")
    spot_files_dir = base_path / 'ST-spotfiles'
    imgs_dir = base_path / 'ST-imgs'
    output_dir = base_path / 'cropped_imgs'

    if not spot_files_dir.is_dir() or not imgs_dir.is_dir():
        print(f"警告: 在 '{base_path}' 中找不到 'ST-spotfiles' 或 'ST-imgs' 文件夹，跳过。")
        return

    output_dir.mkdir(exist_ok=True)
    print(f"裁剪后的图片将保存到: {output_dir}")

    for tsv_path in spot_files_dir.glob('*.tsv'):
        sample_id = tsv_path.stem
        print(f"  - 正在处理样本: {sample_id}")
        
        image_path = imgs_dir / sample_id[0] / f"{sample_id}.jpg"
        if not image_path.exists():
            print(f"    - 警告: 找不到图片 '{image_path}'，跳过。")
            continue

        try:
            df = pd.read_csv(tsv_path, sep='\t')
            source_image = Image.open(image_path)
            width, height = source_image.size

            for index, row in df.iterrows():
                x, y = int(row['x']), int(row['y'])
                pixel_x, pixel_y = int(row['pixel_x']), int(row['pixel_y'])

                half_size = CROP_SIZE // 2
                left, top = pixel_x - half_size, pixel_y - half_size
                right, bottom = pixel_x + half_size, pixel_y + half_size
                
                if not (0 <= left < right <= width and 0 <= top < bottom <= height):
                    print(f"    - 警告: 坐标 ({pixel_x}, {pixel_y}) 裁剪区域超出图像边界，跳过。")
                    continue

                cropped_img = source_image.crop((left, top, right, bottom))
                output_filename = f"{sample_id}_{x}x{y}.jpg"
                cropped_img.save(output_dir / output_filename)

        except Exception as e:
            print(f"    - 错误: 处理 {tsv_path.name} 时发生错误: {e}")

    print(f"--- 数据集 {name} 处理完成 ---\n")

# --- 3. 循环执行 ---
for name, path in datasets.items():
    process_dataset(name, Path(path))

print("所有任务已完成！")
```
