```python
import os
import gzip
import shutil
import re

def delete_file(file_path):
    try:
        if not os.path.isfile(file_path):
            print(f"The file '{file_path}' does not exist.")
            return
        os.remove(file_path)
        print(f"Successfully deleted the file: '{file_path}'.")
    except Exception as e:
        print(f"An error occurred while trying to delete the file: {e}")

def unzip_gz_file(gz_filepath, extract_to_folder):
    if not os.path.exists(extract_to_folder):
        os.makedirs(extract_to_folder)
    output_filepath = os.path.join(extract_to_folder, os.path.basename(gz_filepath)[:-3])
    with gzip.open(gz_filepath, 'rb') as gz_ref:
        with open(output_filepath, 'wb') as out_file:
            shutil.copyfileobj(gz_ref, out_file)
    print(f"Extracted file to {output_filepath}")

gz_pattern = re.compile(r'.*\.gz$')
for file in os.listdir(her2st_gene_expression):
    if gz_pattern.match(file):
        unzip_gz_file(her2st_gene_expression+'/'+file , her2st_gene_expression)
for file in os.listdir(her2st_gene_expression):
    if gz_pattern.match(file):
        delete_file(her2st_gene_expression+'/'+file)
        
```

```python
os.makedirs(stnet_imgs_folder, exist_ok=True)
os.makedirs(stnet_st_postion, exist_ok=True)
os.makedirs(stnet_gene_expression, exist_ok=True)

def move_stdata_gz_files(source_directory, target_directory):
    # Ensure the target directory exists; create it if it doesn't
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)

    # Regular expression to match *_stdata.tsv.gz files
    pattern = re.compile(r'.*_stdata\.tsv\.gz$')

    # Iterate over all files in the source directory
    for filename in os.listdir(source_directory):
        if pattern.match(filename):
            source_file_path = os.path.join(source_directory, filename)
            target_file_path = os.path.join(target_directory, filename)
            
            # Move file to the target directory
            shutil.move(source_file_path, target_file_path)
            print(f"Moved: {source_file_path} to {target_file_path}")

def move_position_gz_files(source_directory, target_directory):
    # Ensure the target directory exists; create it if it doesn't
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)

    # Regular expression to match *_stdata.tsv.gz files
    pattern1 = re.compile(r'.*_Coords\.tsv\.gz$')
    pattern2 = re.compile(r'.*\.csv\.gz$')

    # Iterate over all files in the source directory
    for filename in os.listdir(source_directory):
        if pattern1.match(filename) or pattern2.match(filename):
            source_file_path = os.path.join(source_directory, filename)
            target_file_path = os.path.join(target_directory, filename)
            
            # Move file to the target directory
            shutil.move(source_file_path, target_file_path)
            print(f"Moved: {source_file_path} to {target_file_path}")

def move_img_files(source_directory, target_directory):
    # Ensure the target directory exists; create it if it doesn't
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)

    pattern = re.compile(r'.*.jpg$')

    for filename in os.listdir(source_directory):
        if pattern.match(filename):
            source_file_path = os.path.join(source_directory, filename)
            target_file_path = os.path.join(target_directory, filename)
            
            shutil.move(source_file_path, target_file_path)
            print(f"Moved: {source_file_path} to {target_file_path}")

move_stdata_gz_files(stnet_folder,stnet_gene_expression)
move_position_gz_files(stnet_folder,stnet_st_postion)
move_img_files(stnet_folder,stnet_imgs_folder)
```

```python
import os
import pandas as pd

for file_name in os.listdir(mouse_st_postion):
    csv_path = os.path.join(mouse_st_postion, file_name)
    df = pd.read_csv(csv_path)

    parsed = (
        df['Unnamed: 0']
        .str.extract(r'(\d+(?:\.\d+)?)x(\d+(?:\.\d+)?)$')
        .astype(float)
        .round()
        .astype(int)
    )
    df['x'] = parsed[0]
    df['y'] = parsed[1]

    df = df.rename(columns={'xaxis': 'pixel_x', 'yaxis': 'pixel_y'})
    df = df.drop(columns=['Unnamed: 0'])
    df = df[['x', 'y', 'pixel_x', 'pixel_y']]
    df.to_csv(csv_path, index=False)
```
